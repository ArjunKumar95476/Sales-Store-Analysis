```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # pip install seaborn
```

# Read CSV Data 

```python
df=pd.read_csv('Superstore_Records.csv', encoding='latin-1') # encoding='latin-1' machine can read special character also like /,- etc
```

```python
df.head()
```

## Checking Missing Values

```python
df.info()
```

 #### There is no missing Values In Above Data Set 
 #### But We can see column Order Date and Ship Date containing, Date But Dates are in Object Format So We Have to Change Data Type for
 #### These two Columns

## Converting Date Columns

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
df.head()
```

```python
# Now Adding Three Columns Order Month, Order Year and Order day of Week , bcz I have to do monthly , yearly and Day wise sales 
# Analysis
```

### Adding Columns 

```python
df['Order Month']=df['Order Date'].dt.month
df['Order Year']=df['Order Date'].dt.year
df['Order Day of Week']=df['Order Date'].dt.dayofweek
```

```python
df.head()
```

### 1. You need to calculate the monthly sales of the store and identify which month had the highest sales and which month had the lowest sales.

```python
sales_by_month=df.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
from matplotlib.ticker import FuncFormatter # import for showing value in K format place decimal
# plt.figure(figsize=(8,6)) # plot size
plt.plot(sales_by_month['Order Month'],sales_by_month['Sales'],marker='o')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/1000:.0f}K')) # logic 
plt.title('Monthly sales Analysis')
plt.xlabel('Order Month')
plt.ylabel('Sales(K)')
plt.grid(True)
plt.show()
```

#### Now We  Can Analyze the Data and We Can Say 
#### Highest Sales Month October
#### Lowest Sales Month January

### 2. You need to analyze sales based on product categories and determine which category has the lowest sales and which category has the highest sales.

```python
sales_by_category=df.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
plt.pie(
    sales_by_category['Sales'],
    # explode=[0.1,0.1,0.1],
    labels=sales_by_category['Category'],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    #labels Inside chart
    labeldistance=0.5,
    #percentage inside chart
    pctdistance=0.7
)
plt.title('Sales Analysis By Category ')
plt.axis('equal')
plt.show()
```

#### Highest Sales Category:Furniture and Technology
#### Lowest Sales Category: Office Supplies

### 3. The sales analysis needs to be done based on sub-categories

```python
sales_by_sub_category=df.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_sub_category
```

```python
from matplotlib.ticker import FuncFormatter # import for showing value in K format place decimal
plt.figure(figsize=(12,5))
plt.bar(sales_by_sub_category['Sub-Category'],sales_by_sub_category['Sales'])
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/1000:.0f}K')) # logic y axis in K
plt.xticks(rotation=45, ha='right')
plt.title('Sales Analysis by Sub Category')
plt.xlabel('Sub Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

### 4. You need to analyze the monthly profit from sales and determine which month had the highest profit.

```python
profit_by_monthly=df.groupby('Order Month')['Profit'].sum().reset_index()
```

```python
profit_by_monthly
```

```python
from matplotlib.ticker import FuncFormatter # import for showing value in K format place decimal
# plt.figure(figsize=(8,6)) # plot size
plt.figure(figsize=(12,5))
plt.plot(profit_by_monthly['Order Month'],profit_by_monthly['Profit'],marker='o')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/1000:.0f}K')) # logic 
plt.title('Monthly Profit Analysis')
plt.xlabel('Order Month')
plt.ylabel('Profit(K)')
plt.grid(True)
plt.show()
```

#### Highest Profit Month= October
#### lowest Profit Month= September

### 6. Analyze the sales and profit by customer segment

```python
df.head()
```

```python
sales_profit_by_segment=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
```

```python
sales_profit_by_segment
```

```python
from matplotlib.ticker import FuncFormatter # import for showing value in K format place decimal
sales_profit_by_segment.plot(x='Segment',y=['Sales','Profit'],kind='bar',figsize=(8,4))
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/100000:.0f}M')) # logic 
plt.xticks(rotation=0, ha='right')
plt.title('Sales and Profit by Segment')
plt.xlabel('Segment')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

### 7. Analyze the sales to profit ratio


```python
sales_profit_by_segment=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['sales_to_profit_ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
```
